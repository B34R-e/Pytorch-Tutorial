### **Backpropagation**

#### Chain rule

#### Computational Graph

1. Foward pass: Compute Loss
2. Compute local gradients
3. Backward process: Compute dLoss/dWeights using the Chain Rule

#### Coding

```python
import torch
x = torch.tensor(1.0)
y = torch.tensor(2.0)
w = torch.tensor(1.0, requires_grad = True)
lr = 0.01
epochs = 5

# forward pass
def forward(w, x):
  return x * w
# compute the loss
def compute_loss(y_hat, y):
  return (y_hat - y)**2
new_w = w
for epoch in range(epochs):
  y_hat = forward(w, x)
  loss = compute_loss(y_hat, y)
  print(loss)

  # backward pass
  loss.backward()
  print(w.grad)

  ### update weights
  ### next foward and backward
  new_w = new_w - lr * (w.grad)
  w.grad.zero_()
  print(f'new_weight of epoch {epoch + 1}: {new_w}')
  print(f'loss of epoch {epoch + 1}: {loss}')
```
